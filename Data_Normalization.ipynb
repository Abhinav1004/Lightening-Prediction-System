```python
import pandas as pd
data=pd.read_pickle('saved/improved_entry')
```

```python
from sklearn.preprocessing import MinMaxScaler
x = data[['Temperature Gradient']].values.astype(float)
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data['Temperature Gradient'] = pd.DataFrame(x_scaled)
data.head()
```

```python
x = data[['Pressure']].values.astype(float)
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data['Pressure'] = pd.DataFrame(x_scaled)
data.head()
```

```python
x = data[['Wind Bearing Gradient']].values.astype(float)
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data['Wind Bearing Gradient'] = pd.DataFrame(x_scaled)
data.head()
```

```python
x = data[['Visibility']].values.astype(float)
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data['Visibility'] = pd.DataFrame(x_scaled)
data.head()
```

```python
x = data[['Wind Speed']].values.astype(float)
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data['Wind Speed'] = pd.DataFrame(x_scaled)
data.head()
```
